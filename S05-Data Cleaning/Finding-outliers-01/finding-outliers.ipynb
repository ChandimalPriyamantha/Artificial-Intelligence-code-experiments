-Type of Outliers
  - Univariate Outliers
  - Bivariate Outliers

-Univariate Outliers are the points which are beyond the normal values in a signal variable.

--simple word --> The outliers in a signal variable are  called univeriate outliers.

-Bivariable Outliers are the points which lie far from the expected values when two variables are plotted against others.

Outliers in the data
Outliers are extreme value that fall a long way outdide of the other observations. They may indicate a variability in a meaurement, experimental errors or a novertly.

```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
#Reading the dataset
data = pd.read_csv("bigmart.csv");
```

```python
data.head()
```

```python
#Checking for univariate outliers

plt.rcParams['figure.figsize'] = (16,4)

plt.subplot(1,4,1)
sns.boxplot(data['Item_Weight'])

plt.subplot(1,4,2)
sns.boxplot(data['Item_Visibility'])

plt.subplot(1,4,3)
sns.boxplot(data['Item_MRP'])

plt.subplot(1,4,4)
sns.boxplot(data['Item_Outlet_Sales'])

plt.suptitle("Checking for Univeriate Outliers")
plt.show()
```

```python
#Checking for the bivariate outliers wrt Target variable.
plt.subplot(1,3,1)
sns.scatterplot(y=data['Item_Outlet_Sales'],x=data['Item_Weight'])

plt.subplot(1,3,2)
sns.scatterplot(y=data['Item_Outlet_Sales'],x=data['Item_Visibility'])

plt.subplot(1,3,3)
sns.scatterplot(y=data['Item_Outlet_Sales'],x=data['Item_MRP'])

plt.suptitle("Checking for Bivariate Outliers")
plt.show()
```

-Handiling Outliers

Example:
  - The age variable in our dataset which predicts loan approval
  - The normal range of age can be 15-60 years but there may be some people with age
    greater than 60.
  - So in this case, where people having age greater than 60 applied for the loan
  are not outliers, we can not simply delete them.

#Deleting Outliers

-We are sure that the outliers are due to an entry error or due to measurment error.
-If the outliers create a siginificant relationship between two independent variables which is
 against the assumption of many of our machine learing algoritums.

#Capping Outliers

- Capping refers to replacing the outliers to a near value so that we can keep the point
in our analysis and it also does not skew the data.
#Note: Other than Deletind and Capping there are two more ways of handing outliers.
     1. Imputing: If an outlier seems to be due to some mistake and we recognise the mistake.
     2. Binning: It is the process of transorming numerical variables into categorical type.
        Example: we can bin the age variable into categories such as 20-40, 40-60, 60-90 and above.

-Capping outliers
-Capping refers to replacing the outliers to a near value so that we can keep the point in our analysis and it also does not skew the data.

```python
#Capping the outliers in the colum item_outlet_sales having values greater than 10000
data['Item_Outlet_Sales'].values[data['Item_Outlet_Sales'].values>10000]=10000
```

```python
#Checking if there is still any value greater than 10000 left in the Item_Outlet_Sales colum
data[data['Item_Outlet_Sales']>10000]
```

```python
data = data[~(data['Item_Outlet_Sales']>8000)]
```
